# Probability

```python
import random
from collections import Counter

def roll_dice(n):
    result = []
    for i in range(n):
        result.append(random.randint(1,6))
    return result

dice_result = roll_dice(6)

print("result : ", dice_result)
```

```python
def counter_occurrences(results):
    return Counter(results) # results에 있는 중복된 함수 찾기 : Counter 함수

def calculate_probabilities(counts,n):
    probabilities={}
    
    for num in range(1,7):
        probabilities[num] = counts[num] / n
        
    return probabilities
    
num_trails=100000

dice_result = roll_dice(num_trails)
counts = counter_occurrences(dice_result)
probabilities = calculate_probabilities(counts,num_trails)
print("각 숫자가 나온 확률: ")
for number, prob in probabilities.items():
    print(f"{number}: {prob:.3f}")
```

# Conditional Probability

```python
# 조건부 확률 계산 함수
def conditional_probability(event_count, condition_count):
    if condition_count == 0:
        return 0
    return event_count / condition_count

#데이터 정의

students = 10
male = 5
female = 5 
male_with_lap = 2
female_with_lap = 3

# 확률 계산
plwm = conditional_probability(male_with_lap, male)
plwf = conditional_probability(female_with_lap,female)

print(f"p(노트북|남학생)={plwm:.2f}")
print(f"p(노트북|여학생)={plwf:.2f}")
```
